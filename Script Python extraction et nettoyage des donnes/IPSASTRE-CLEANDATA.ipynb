```python
ipsastre=ipsastre.replace(to_replace ="Prépa intégrée",
                 value ="Prepa integree")
```

```python
ipsastre=ipsastre.replace(to_replace ="Prépa classique",
                 value ="Prepa classique")
```

```python
ipsastre=ipsastre.replace(to_replace ="Prépa ATS",
                 value ="Prepa ATS")
```

```python
ipsastre=ipsastre.replace(to_replace ="Plutôt milieu",
                 value ="Plutot milieu")
```

```python
ipsastre=ipsastre.replace(to_replace ="Plutôt derrière 	",
                 value ="Plutot derriere")
```

```python
ipsastre
```

```python
def remove_e(text):
    pattern = r'[^0-9.,!?/:;\"\'\s]' 
    return re.sub(pattern, '', text)
```

```python
ipsastre['numEtu'] = ipsastre.numEtu.apply(func = remove_e)
```

```python
ipsastre
```

```python
def non_ascii(s):
  return "".join(i for i in s if ord(i)<128)
```

```python
ipsastre['monterPc'] = ipsastre.monterPc.apply(func = non_ascii)
```

```python
ipsastre['assosEtu'] = ipsastre.assosEtu.apply(func = non_ascii)
```
